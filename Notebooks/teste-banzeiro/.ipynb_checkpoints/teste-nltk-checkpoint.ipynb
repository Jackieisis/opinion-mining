```python
import nltk
import json
import emoji
```

```python
with open("banzeiro.json") as file:
    publications = json.load(file)
with open("banzeiro.json") as file:
    publicationsEmoji = json.load(file)
```

```python
for p in publications:
    if "comments" in p:
        for c in p["comments"]:
            c["comment"] = emoji.demojize(c["comment"])
```

```python
nltk.download("punkt")
```

```python
sent_tokenizer=nltk.data.load('tokenizers/punkt/portuguese.pickle')

for p in publications:
    if "caption" in p:
        print("\n--------------------------------------\n", p["caption"])
        caption_sentence = sent_tokenizer.tokenize(p["caption"])
        print("<<", caption_sentence, ">>")
    if "comments" in p:
        for c in p["comments"]:
            print ("\n ->", c["comment"], "(por: "+c["author"]+")")
            comment_sentence = sent_tokenizer.tokenize(c["comment"])
            print("<<", comment_sentence, ">>")
```
